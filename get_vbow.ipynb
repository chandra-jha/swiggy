```python
def get_vbow(sample_path, scaler, pca, model):
    x1 = np.loadtxt(sample_path, delimiter=',')
    x1_norm = scaler.transform(x1)
    x1_norm_pca = pca.transform(x1_norm)
    x1_pred = model.predict(x1_norm_pca)
    x1_pred_hist = np.zeros(100)
    for v in x1_pred:
        x1_pred_hist[v] += 1
    x1_pred_hist /= x1_pred_hist.sum()
    name = os.path.basename(sample_path).strip(".feat")
    np.savetxt("/home/ubuntu/foodissues_vbow/" + name + ".vbow", x1_pred_hist, delimiter = ',')
```

```python
sift_dir = "/home/ubuntu/foodissues_features/"
sift_list = glob(sift_dir + "/*")

count = 0
for s in sift_list:
    name = os.path.basename(s).strip(".feat")
    if not os.path.exists("/home/ubuntu/foodissues_vbow/" + name + ".vbow"):
        try:
            get_vbow(s, scaler, pca, model)
            count += 1
        except:
            pass
    else:
        count += 1
    if count % 50 == 0:
        print(count)
```
